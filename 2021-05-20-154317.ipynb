```r
x<-c(16,18.5,22,27,32,37,42)
y<-c(34,86.5,111.1,113.9,84.5,35.4,6.8)
relation<-lm(y~x)
print(relation)
summary(relation)
```

```r
plot(x,y,col="blue", main="Linear Regression Example",abline(lm(y~x)), cex=1.3,pch=16,xlab="x", ylab="y")
dev.off()
```

```r
confint(relation, level=0.95)
```

```r
plot(fitted(relation), residuals(relation))
```

```r
x<-c(16,18.5,22,27,32,37,42)
y<-c(34,86.5,111.1,113.9,84.5,35.4,6.8)
relation2<-lm(y~poly(x,2))
print(relation2)
summary(relation2)
```

```r
x_seq <- seq(from=16,to=42, by=0.1)
new_relation2<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation2[i]<-predict(relation2, x_new)
}
plot(x,y,col="blue", main="Linear Regression Example",lines(x_seq, new_relation2),cex=1.3,pch=16,xlab="x",ylab="y")
```

```r
confint(relation2, level=0.95)
```

```r
plot(fitted(relation2), residuals(relation2))
```
